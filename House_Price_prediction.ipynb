This will predict the house prices based on the given input.


```python
import pandas as pd #impoting Libraries
```

```python
df=pd.read_csv('House_price - House_price.csv') # reading file
```

```python
df #Displaying the dataset
df=df.dropna()
```

```python
x=df[['Number of Bedrooms']] #selecting variables
```

```python
y=df[['Price']]
```

```python
import matplotlib.pyplot as plt
```

```python
#to plot the points in graph
plt.scatter(x,y)
plt.plot(x,y)#to draw the lines in graphs
plt.show()#to  show the values
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pre=lr.predict(x_test)
```

```python
y_pre
```

```python
y_test.head(),y_pre[0:5] #comparing values
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("The Mean Square Error is:",mean_squared_error(y_test,y_pre))
```

```python
plt.scatter(x_train,y_train)#scattering the values of train
plt.plot(x_test,y_pre)#Drawing the line of x_test and y_pre
plt.show()#showing the graph
```
